```python
print("Numerical columns")
for col in df.columns:
  if df[col].dtypes !="object":
    print(col)
print("Categorical columns")
for col in df.columns:
  if df[col].dtypes =="object":
   print(col)
```

```python
ptl.subplot(2,2,1)
sns.histplot(df["age"])
ptl.title("age distribution")
ptl.xlabel("age")
ptl.ylabel("count")

ptl.subplot(2,2,2)
sns.histplot(df["sleep_hours"])
ptl.title("sleep_hours")
ptl.xlabel("sleep")
ptl.ylabel("count")

ptl.subplot(2,2,3)
sns.histplot(df["addiction_level"])
ptl.title("addiction_level")
ptl.xlabel("addiction")
ptl.ylabel("count")

ptl.subplot(2,2,4)
sns.histplot(df["anxiety_level"])
ptl.title("anxiety_level")
ptl.xlabel("anxiety_level")
ptl.ylabel("count")
ptl.show()
ptl.tight_layout()


```

```python
ptl.subplot(2,2,1)
sns.countplot(x="gender",data=df)
ptl.title("gender count")
ptl.xlabel("gender")
ptl.ylabel("count")


ptl.subplot(2,2,2)
sns.countplot(x="platform_usage",data=df)
ptl.title("platform_usage count")
ptl.xlabel("platform_usage")
ptl.ylabel("count")



```

```python
correlated=df.corr(numeric_only=True)
sns.heatmap(correlated,annot=True)
```

```python
#average of daily_social_media_hours
print("Average=",df["daily_social_media_hours"].mean())
#maximun value of  expenses
print("Maximum value=",df["daily_social_media_hours"].max())
#minimun value of  expenses
print("Minimum value=",df["daily_social_media_hours"].min())
#average expenses by smoker
print("average daily_social_media_hours by sleep_hours=",df.groupby("sleep_hours")["daily_social_media_hours"].mean())
```

```python
sns.boxplot(x="gender",y="platform_usage",hue="social_interaction_level",data=df)
ptl.title("Social media addiction")
ptl.xlabel("gender")
ptl.ylabel("addiction")
ptl.show()
```
